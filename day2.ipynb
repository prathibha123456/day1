```python
#methods
# append(), insert(), extend()
fruits = ["apple", "banana"]
#append()--add to the END
fruits.append("cherry")
print(fruits)
```

```python
fruits = ["apple","banana"]
fruits.insert(3,"cherry")
print(fruits)
```

```python
fruits = ["apple","banana"]
fruits.extend(["kiwi"])
print(fruits)
```

```python
#   Removing Element remove()
fruits = ["apple","banana","cherry","mango"]
fruits.remove("banana")
print(fruits)
```

```python
#pop(0)---remove first element
fruits = ["apple","banana","mango"]
fruits.pop(0)
```
